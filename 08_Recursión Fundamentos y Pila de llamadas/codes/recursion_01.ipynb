# Generación de Subconjuntos: Explorando el Árbol de Decisiones

Este bloque de código resuelve un problema clásico: generar el conjunto potencia (todos los subconjuntos posibles) de un conjunto dado, en este caso $\{1, 2, 3\}$. A diferencia de la recursión lineal vista en secuencias como Fibonacci, aquí utilizamos recursión de ramificación para explorar múltiples caminos simultáneos. Para entender este algoritmo, no debemos pensar en una línea recta, sino en un Árbol de Espacio de Estados (State Space Tree). Por cada elemento del conjunto original, tomamos una decisión binaria: ¿Lo incluimos en nuestro subconjunto o lo excluimos?

## Análisis Paso a Paso del Código

### 1. La Firma de la Función y el Manejo de Memoria

```cpp
void generarSubconjuntos(const std::vector<int>& conjunto, int indice, std::vector<int>& subconjuntoActual)

El paso de parámetros es fundamental para la eficiencia del algoritmo:

* `const std::vector<int>& conjunto`: Pasamos el conjunto original por referencia constante. Solo necesitamos leerlo, y al evitar pasarlo por valor, nos ahorramos copiar el arreglo entero en cada marco de la pila de llamadas (Stack Frame).

* `std::vector<int>& subconjuntoActual`: Este es nuestro estado mutable. Lo pasamos por referencia para que todas las llamadas recursivas operen sobre el mismo espacio de memoria subyacente, en lugar de crear un vector nuevo en cada paso.

### 2. El Caso Base (Condición de Parada)

```cpp
if (indice == conjunto.size()) { ... return; }

Llegamos al caso base cuando el indice alcanza el tamaño del conjunto original (en este caso, 3). Esto significa que hemos tomado una decisión (incluir/excluir) para cada uno de los elementos. En este punto, el estado de subconjuntoActual representa una combinación válida y la imprimimos. El return es vital para detener la recursión y comenzar a desapilar.

### 3. La Ramificación (Exploración de Caminos)
El corazón de la recursión divide la ejecución en dos universos paralelos:

* **Rama A (Excluir)**: `generarSubconjuntos(conjunto, indice + 1, subconjuntoActual);`
Avanzamos al siguiente índice sin modificar el estado actual. Decidimos no incluir el elemento actual en el subconjunto.

* **Rama B (Incluir)**: `subconjuntoActual.push_back(conjunto[indice]);`
Modificamos el estado agregando el elemento actual al vector y luego hacemos la llamada recursiva para avanzar.

### 4. El Paso de "Backtracking" (Limpieza del Estado)

```cpp
subconjuntoActual.pop_back();

Esta es la línea más sutil y poderosa del código. Cuando la "**Rama B**" termina de ejecutarse y la función retorna, el marco de la pila se destruye. Sin embargo, como subconjuntoActual fue pasado por referencia, el elemento que añadimos con push_back sigue ahí.Para que la ejecución pueda volver atrás en el árbol y explorar otros caminos correctamente, debemos deshacer nuestro cambio. El pop_back() restaura el vector a su estado original antes de que tomáramos la decisión de "Incluir".
#### Complejidad Computacional
* Para evaluar la viabilidad de este algoritmo, analizamos su comportamiento asintótico:Complejidad Temporal: Por cada elemento, tomamos 2 decisiones. Para un conjunto de tamaño $n$, el árbol tiene $2^n$ hojas. En cada hoja iteramos sobre el subconjunto (que tiene en promedio tamaño $n/2$) para imprimirlo. Por lo tanto, la complejidad temporal es $O(n \cdot 2^n)$. Es un algoritmo exponencial, inherente a la naturaleza de generar un conjunto potencia.
* Complejidad Espacial: Gracias a que pasamos los vectores por referencia, la memoria adicional requerida está dictada únicamente por la profundidad máxima de la pila de llamadas. Esta profundidad es exactamente igual al número de decisiones que tomamos en un solo camino, es decir, $n$. La complejidad espacial es $O(n)$.

# Código

````c++
#include <iostream>
#include <vector>

// Función recursiva para generar subconjuntos
void generarSubconjuntos(const std::vector<int>& conjunto, int indice, std::vector<int>& subconjuntoActual) {
    // 1. Caso Base: Hemos tomado una decisión para todos los elementos
    if (indice == conjunto.size()) {
        std::cout << "{ ";
        for (int num : subconjuntoActual) {
            std::cout << num << " ";
        }
        std::cout << "}\n";
        return;
    }

    // 2. Rama A: EXCLUIR el elemento actual y avanzar
    generarSubconjuntos(conjunto, indice + 1, subconjuntoActual);

    // 3. Rama B: INCLUIR el elemento actual y avanzar
    subconjuntoActual.push_back(conjunto[indice]); // Modificamos el estado
    generarSubconjuntos(conjunto, indice + 1, subconjuntoActual);

    // 4. BACKTRACKING (Limpieza del estado): 
    // Al regresar (desapilar), debemos restaurar el vector a su estado original
    // para que la otra rama no se contamine con este elemento.
    subconjuntoActual.pop_back(); 
}
````
