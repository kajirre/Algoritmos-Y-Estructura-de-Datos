# Recursión 
Todo algoritmo recursivo debe cumplir con dos reglas de oro para ser funcional y evitar el desbordamiento de la pila (stack overflow):
* **Caso Base**: La condición de parada que devuelve un valor sin realizar más llamadas recursivas.
* **Paso Recursivo**: La llamada a la función con un argumento que acerca el estado hacia el caso base.Ejemplo de Código: Factorial de un númeroPara ilustrar la recursión matemática simple, utilizaremos el cálculo del factorial ($n!$), que se define como $n \times (n-1)!$, siendo $0! = 1$.

````c++
#include <iostream>

/**
 * Calcula el factorial de un número de forma recursiva.
 * @param n Número entero no negativo.
 * @return El factorial de n.
 */
unsigned long long factorial(int n) {
    // 1. Caso Base: Si n es 0 o 1, el factorial es 1.
    // Esto evita que la función se llame infinitamente.
    if (n <= 1) {
        return 1; 
    }
    
    // 2. Paso Recursivo: n! = n * (n-1)!
    // La función se llama a sí misma con un problema más pequeño (n-1).
    return n * factorial(n - 1);
}
````

````c++
int num = 5;
std::cout << "El factorial de " << num << " es: " << factorial(num) << std::endl;
````

El caso base `if (n <= 1)` es fundamental. Sin él, el programa intentaría calcular el factorial de números negativos indefinidamente, agotando la memoria de la pila de llamadas (call stack).

En la la línea `return n * factorial(n - 1);` ejecución actual se "pausa" y se abre un nuevo marco de pila (stack frame) para `factorial(n - 1)`. Una vez que el caso base responde, las multiplicaciones se resuelven en orden inverso (de abajo hacia arriba).

## Ejemplo generar binarios 

Sea $P(n)$ la proposición: "El algoritmo genera todas las $2^n$ secuencias binarias únicas de longitud $n$".

### Demostración por Inducción
* **Caso Base $P(0)$**: Si $n = 0$, la condición actual.length() == 0 se cumple de inmediato. El algoritmo imprime una cadena vacía (la única secuencia de longitud 0) y termina. $P(0)$ es verdadero.
* **Hipótesis Inductiva**: Supongamos que para un valor $k$, la función generarBinario es capaz de generar correctamente todas las secuencias de longitud $k$.
* **Paso Inductivo $P(k+1)$**:  Al llamar a la función para buscar longitud $k+1$, la primera instrucción recursiva fija el primer bit como '0' y delega el resto del trabajo.
    * Por nuestra hipótesis, la llamada para completar los $k$ bits restantes funcionará correctamente, generando $2^k$ secuencias que empiezan por '0'.
    * Al retornar (limpieza del stack), la segunda instrucción fija el bit como '1' y vuelve a delegar. Esto genera otras $2^k$ secuencias únicas.
    * Total: $2^k + 2^k = 2^{k+1}$ secuencias. Como todas las de la primera rama empiezan con '0' y las de la segunda con '1', garantizamos que no hay duplicados.

````c++
#include <iostream>
#include <string>

/**
 * Genera el espacio de estados binarios de n bits.
 * @param n Longitud total deseada (Constante en la recursión).
 * @param actual Estado acumulado en el Stack Frame.
 */
void generarBinario(int n, std::string actual) {
    // 1. CASO BASE (Condición de parada)
    // Matemáticamente equivale a P(0) o al límite de la inducción.
    if (actual.length() == n) {
        std::cout << "[" << actual << "]" << std::endl;
        return; 
    }

    // 2. PASO RECURSIVO (Hipótesis Inductiva)
    // Se asume que la función sabe resolver el problema para (n - actual.length() - 1)
    generarBinario(n, actual + "0"); // Rama de decisión 0
    generarBinario(n, actual + "1"); // Rama de decisión 1
}
````

* El if actúa como guardián de la memoria. Si se omite, el sistema intentaría crear Stack Frames infinitos, provocando un Segmentation Fault.
* `generarBinario(n, actual + "0");` y `generarBinario(n, actual + "1");` Representan la bifurcación del universo de cómputo. En C++, la cadena actual se pasa por valor (o se crea una nueva), lo que significa que cada nivel de la recursión mantiene su propia copia del "prefijo", permitiendo que al regresar de la llamada `generarBinario(n, actual + "0");`, el estado para la `generarBinario(n, actual + "1");` sea idéntico al original.

````c++
int bits = 1;
std::cout << "Espacio de estados para " << bits << " bits:" << std::endl;
generarBinario(bits, "");
````

## El Backtracking

La recursión simple (como el factorial o Hanói) sigue un camino lineal o una ramificación fija. Sin embargo, en el mundo real, los problemas a menudo requieren toma de decisiones bajo incertidumbre.

El Backtracking (Vuelta Atrás) es una técnica algorítmica que utiliza la recursión para explorar sistemáticamente todas las configuraciones posibles de un espacio de búsqueda. Su esencia es el ensayo y error inteligente:

1. Explora una opción.
2. Verifica si esa opción conduce a una solución.
3. Si no es así (llegamos a un callejón sin salida), retrocede (backtrack) al estado anterior, deshaciendo los cambios realizados, y prueba la siguiente opción.

````c++
#include <vector>
/**
 * Intenta encontrar un camino hacia el objetivo.
 * Ilustra la estructura: Intentar -> Validar -> Retroceder.
 */
bool buscarCamino(std::vector<int>& mapa, int pos, int objetivo) {
    // 1. Caso Base: Éxito
    if (pos == objetivo) {
        return true;
    }

    // 2. Caso Base: Fracaso (Fuera de límites o pared)
    if (pos < 0 || pos >= mapa.size() || mapa[pos] == -1 || mapa[pos] == 2) {
        return false;
    }

    // Marcar como "visitado" temporalmente para evitar ciclos infinitos
    mapa[pos] = 2; 
    std::cout << "Visitando posicion: " << pos << std::endl;

    // Intentar mover a la derecha
    if (buscarCamino(mapa, pos + 1, objetivo)) return true;

    // Intentar mover a la izquierda
    if (buscarCamino(mapa, pos - 1, objetivo)) return true;

    // --- PASO DE BACKTRACKING ---
    // Si ninguna de las llamadas anteriores funcionó, este camino es fallido.
    // "Limpiamos" nuestra marca para dejar el mapa como estaba.
    std::cout << "Retrocediendo desde: " << pos << std::endl;
    mapa[pos] = 0; 
    
    return false;
}
````

````c++
std::vector<int> mapa = {0, -1, 0, 0, 0}; // El -1 bloquea el camino directo
int inicio = 0, meta = 4;

if (buscarCamino(mapa, inicio, meta))
    std::cout << "¡Objetivo alcanzado!" << std::endl;
else
    std::cout << "No hay salida." << std::endl;
````

## El Algoritmo del Laberinto 
En el contexto de una cuadrícula de $N \times M$, el problema de encontrar un camino se define como la navegación a través de un espacio de estados. Cada estado es una coordenada $(x, y)$.

### El Algoritmo
El procedimiento sigue una estrategia de búsqueda en profundidad con retroceso sistemático. Los pasos lógicos son:
1. **Evaluación de Éxito (Caso Base)**: Se comprueba si la posición actual $(x, y)$ coincide con la meta. Si es así, se marca como parte de la solución y se notifica el éxito a la llamada anterior.
2. **Filtro de Inviabilidad (Poda)**: Antes de expandir el estado, se validan tres condiciones críticas:
    1. Límites: ¿Está la coordenada dentro de los rangos $[0, N-1]$ y $[0, M-1]$?
    2. Obstrucción: ¿Es la celda una pared o un obstáculo infranqueable?
    3. Ciclo: ¿Ha sido esta celda visitada ya en la rama de exploración actual? (Evita recursión infinita).
3. **Marcado de Estado**: Se registra la celda actual como "ocupada" o "visitada". Esto modifica el estado global para las llamadas recursivas subsiguientes.
4. **Expansión de Ramas (Recursión)**: Se invoca el algoritmo de forma recursiva para los estados vecinos (típicamente arriba, abajo, izquierda y derecha). Debido a la naturaleza de la pila, estas llamadas se ejecutan de forma secuencial, no simultánea.
    1. **Propagación de Resultados**: Si alguna llamada vecina devuelve éxito, la función actual termina inmediatamente devolviendo éxito (cortocircuito).
5. **Paso de Backtracking (Restauración)**: Si todas las llamadas vecinas devuelven fracaso, se debe deshacer el marcado del punto 3 (limpiar el rastro). La función devuelve fracaso a su predecesor.


```cpp
ALGORITMO ResolverLaberinto(x, y)
    
    // 1. EVALUACIÓN DE ÉXITO (Caso Base de Parada)
    SI (x, y) es igual a la Posición_Meta:
        MARCAR (x, y) como parte de la RUTA_SOLUCIÓN
        RETORNAR Verdadero

    // 2. FILTRO DE INVIABILIDAD (Poda y Validación)
    SI (x, y) está fuera de los límites de la matriz O
       (x, y) es un Obstáculo/Pared O
       (x, y) ya ha sido visitado en la rama actual:
        RETORNAR Falso

    // 3. MARCADO DE ESTADO (Exploración)
    MARCAR (x, y) como VISITADO

    // 4. EXPANSIÓN DE RAMAS (Recursión en 4 Direcciones)
    // Intentamos cada dirección de forma secuencial
    SI ResolverLaberinto(x + 1, y) == Verdadero: RETORNAR Verdadero (Mover Abajo)
    SI ResolverLaberinto(x, y + 1) == Verdadero: RETORNAR Verdadero (Mover Derecha)
    SI ResolverLaberinto(x - 1, y) == Verdadero: RETORNAR Verdadero (Mover Arriba)
    SI ResolverLaberinto(x, y - 1) == Verdadero: RETORNAR Verdadero (Mover Izquierda)

    // 5. PASO DE BACKTRACKING (El "Arrepentimiento")
    // Si llegamos aquí, ninguna de las 4 direcciones llevó a la meta.
    DESMARCAR (x, y) como VISITADO (Restaurar el estado original)
    
    RETORNAR Falso (Informar al nivel anterior que este camino falló)
FIN ALGORITMO
